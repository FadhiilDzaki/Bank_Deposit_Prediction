# Library

```python
import pickle
import pandas as pd
import numpy as np
```

# Data Baru

```python
# Membuat inference
data = {
    'age': [35, 45, 29, 54, 22],
    'job': ['management', 'blue-collar', 'technician', 'retired', 'student'],
    'marital': ['married', 'single', 'single', 'married', 'single'],
    'education': ['tertiary', 'secondary', 'tertiary', 'primary', 'unknown'],
    'default': ['no', 'no', 'no', 'no', 'no'],
    'balance': [5000, 2000, -1000, 1500, 0],
    'housing': ['yes', 'no', 'yes', 'yes', 'no'],
    'loan': ['no', 'no', 'no', 'yes', 'no'],
    'contact': ['cellular', 'telephone', 'cellular', 'telephone', 'cellular'],
    'day': [20, 15, 5, 10, 25],
    'month': ['may', 'jul', 'jun', 'apr', 'sep'],
    'duration': [1200, 900, 300, 450, 1000],
    'campaign': [3, 5, 1, 2, 1],
    'pdays': [90, 30, -1, 60, -1],
    'previous': [2, 1, 0, 0, 0],
    'poutcome': ['success', 'failure', 'unknown', 'other', 'unknown'],
    'deposit': ['yes', 'no', 'yes', 'no', 'yes']
}

# to dataframe
df = pd.DataFrame(data)

# cek df
df
```

```python
# inisiasi list untuk menampung week
week = []

# loop deteksi minggu berdasarkan tanggal
for i in df['day']:
    # kondisi tanggal 1-7
    if i <= 7:
        # add minggu ke-1 ke list
        week.append(1)
    # kondisi tanggal 8-14
    elif i <= 14:
        # add minggu ke-2 ke list
        week.append(2)
    # kondisi tanggal 15-21
    elif i <= 21:
        # add minggu ke-3 ke list
        week.append(3)
    # kondisi tanggal 22-31
    elif 21 < i <= 31:
        # add minggu ke-4 ke list
        week.append(4)
    # kondisi lainnya
    else:
        # add NA ke list0
        week.append(pd.NA)

# add 'week' to dataframe
df['week'] = week

df
```

# Load Model

```python
# load
with open('knn_best.pkl','rb') as file_1:
    knn = pickle.load(file_1)
```

# Predict

```python
# predict
knn.predict(df)
```

Hasil prediksi data inference berturut-turut adalah no, yes, no, no, yes.
